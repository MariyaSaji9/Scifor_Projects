Given 5 categories as snacks, milk, fruits, vegetables, and flour and their respective production values for a company create a comparison visualization to show various aspects of the production for the ABC company in the year 2022

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cat = ['snacks', 'milk', 'fruits', 'vegetables', 'flour']
prod = [560000, 373899, 509499, 2728389, 466777]
plt.bar(cat, prod)
plt.title("production of ABC company in 2022")
plt.xlabel("Categories")
plt.ylabel("Production")
plt.show()
```

given 5 cities in india and their respective population in 2023. plot the population using bar(vertical, horizontal) plots.

```python
cities = ['Kochi', 'Kozhikode', 'Chennai', 'Banglore', 'Mumbai']
population = [560000000, 3738990000, 509499000, 2728389000, 4667770000]
plt.bar(cities, population)
plt.title("Population of 5 cities in India (2023)")
plt.xlabel("Cities")
plt.ylabel("Population")
plt.show()
```

```python
plt.barh(cities, population)
plt.title("Population of 5 cities in India (2023)")
plt.ylabel("Cities")
plt.xlabel("Population")
plt.show()
```
