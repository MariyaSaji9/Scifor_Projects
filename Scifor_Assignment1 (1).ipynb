1. Create a Series Object using ndarray having values as the multiples of 100 in the first 10000
values.


```python
pd.set_option('display.max_rows', None)
arr = np.array([i * 100 for i in range(10000)])
ser = pd.Series(arr)
ser
```

2. Create a DataFrame containing two dictionary Category and Sub-category which consists
values as the various options available in the shopping website along with their quantity.

```python
category = {"Category": ["Electronics", "Clothing", "Books", "Furniture"],"Quantity": [500, 800, 300, 200],}
sub_category = {"Category": ["Laptops", "Saree", "Fiction", "Table"],"Quantity": [200, 400, 150, 100],}
df1 = pd.DataFrame(category)
df2 = pd.DataFrame(sub_category)
df = pd.concat([df1, df2], axis=1)
df.columns = ["Category", "Category_Quantity", "Sub_Category", "Sub_Quantity"]
df
```
