3. Create the channels of rgb for the image.

```python
# Split the image into its individual color channels
b, g, r = cv2.split(img)
cv2_imshow(b)
cv2_imshow(g)
cv2_imshow(r)
# Wait for a key press and close all OpenCV windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```

4. code to make a line and arrowedline of a specified img

```python
#line
start_point = (10, 10)
end_point = (200, 200)
#color in BGR format (Blue, Green, Red)
color = (0, 0, 255)
# Draw line
image_with_line = cv2.line(img, start_point, end_point, color, thickness=2)
cv2_imshow(image_with_line)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#arrowed line
start_point_arrow = (10, 10)
end_point_arrow = (200, 150)
# Draw arrowed line
image_with_arrowed_line = cv2.arrowedLine(img, start_point_arrow, end_point_arrow, color,thickness = 5, tipLength=0.3)
cv2_imshow(image_with_arrowed_line)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

6. draw the circle, ellipse, put text on the image, draw rectangle, rotate the image

```python
import numpy as np
cv2.circle(img, (150, 150), 50, (255, 0, 0), thickness=2)
cv2.ellipse(img, (110, 110), (100, 50), 0, 0, 360, (0, 255, 255), thickness=2)
text = "Some Drawings"
cv2.putText(img, text, (50, 300), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), thickness=2)
cv2.rectangle(img, (150, 50), (200, 150), (0, 0, 255), thickness=2)
cv2_imshow(img)
cv2.waitKey(0)
rotated_img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
