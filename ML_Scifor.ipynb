```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

crt new csv file, sale of production company(4 years data)(different type of food). plot the dataframe. display the data from one year to another year. modify dataset by adding columns.

```python
df = pd.read_csv("/content/product.csv")
```

```python
df
```

```python
plt.figure(figsize = (10,6))
sns.lineplot(data = df, x ='product', y = 'Y_2010')
plt.xlabel("Product")
plt.xlabel("Year")
plt.legend()
plt.title("Production grapgh according to years")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.barplot(data = df, x ='product', y = 'Y_2011')
plt.xlabel("Product")
plt.xlabel("Year")
plt.legend()
plt.title("Production grapgh according to years")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(data = df, x ='product', y = 'Y_2012')
plt.xlabel("Product")
plt.xlabel("Year")
plt.legend()
plt.title("Production grapgh according to years")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.pointplot(data = df, x ='product', y = 'Y_2013')
plt.xlabel("Product")
plt.xlabel("Year")
plt.legend()
plt.title("Production grapgh according to years")
plt.show()
```
