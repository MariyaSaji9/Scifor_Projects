```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {'Name of Student':['Ammu', 'Appu', 'Achu', 'Akshay', 'Anas', 'Arathy', 'Aswathy', 'Amar', 'Basil', 'Amritha', 'Malu', 'Minnu', 'Chinnu', 'Chikku', 'Chinju' ], 'Course':['CS', 'IT', 'AI', 'DS', 'AI', 'DS', 'IT', 'AI', 'CS', 'IT', 'AI', 'DS', 'AI', 'DS', 'IT'], 'Place': ['kochi', 'kollam', 'kochi','malappuram', 'chennai', 'coim', 'kochi', 'kollam', 'kochi','malappuram', 'chennai', 'coim', 'pala', 'palakkad', 'kochi'], 'marks':[98, 97,95,96,95,96,97,94,96,94,97,95,96,96,97], 'grade':['A', 'A','B','A','B','A', 'A','B','A','B','A', 'A','B','A','B'],'Age': [18, 19, 20, 21, 22, 23, 24, 18, 19, 20, 21, 22, 23, 24, 18],'Gender': ['FeMale', 'male', 'Male', 'male', 'Male', 'Female', 'feMale', 'male', 'Male', 'Female', 'feMale', 'Female', 'feMale', 'Female', 'feMale']}
df = pd.DataFrame(data)
df
```

```python
df.to_csv("student_record.csv")
```

```python
plt.figure(figsize=(8, 6))

sns.boxplot(data=df, x="Gender", y="Age")
plt.title("Age Distribution by Gender")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Name of Student', y='Age')
plt.title('Scatter Plot of Age vs. Student Name')
plt.xlabel('Student Name')
plt.ylabel('Age')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

sns.violinplot(data=df, x='grade', y='marks')
plt.title('Violin Plot of marks by Grade')
plt.xlabel('Grade')
plt.ylabel('marks')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

sns.pointplot(data=df, x='grade', y='Age')
plt.title('Point Plot of Average Age by Grade')
plt.xlabel('Grade')
plt.ylabel('Average Age')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

sns.lineplot(data=df, x='Name of Student', y='Age')
plt.title('Line Plot of Age vs. Student Name')
plt.xlabel('Student Name')
plt.ylabel('Age')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

sns.barplot(data=df, x='Age')
plt.title('Bar Plot of Student Count by Grade')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

sns.histplot(data = df, x='Age', kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
