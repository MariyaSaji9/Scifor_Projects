```python
#import necessary libraries
import gym #for RL environment
from stable_baselines3 import PPO #proximal policy optimization algorithm
from stable_baselines3.common.vec_env import DummyVecEnv #wrapper for vectorized environment
from stable_baselines3.common.evaluation import evaluate_policy #for evaluation
```

Testing the environment

```python
#defining the environment and creating the gym environment with new_step_api
environment_name = 'LunarLander-v2'
env = gym.make(environment_name, new_step_api=True)
```

```python
#number of episodes for testing
episodes = 10
for episode in range(1, episodes + 1):
    #Reset the environment for each episode and retrieve the initial state
    state = env.reset()
    done = False  #track if the episode is finished
    score = 0  #Initialize the score for the episode
    while not done:
        env.render()  #Render the environment (display the game)
        action = env.action_space.sample()  #Sample a random action from the action space
        step_result = env.step(action)  #Take a step in the environment by performing the sampled action
        print("Step result:", step_result)
        score += step_result[1] #Accumulate the reward received from the step
        done = step_result[2]  #Update the 'done' flag
        print('Episode:{} Score:{}'.format(episode, score))
env.close()
```
