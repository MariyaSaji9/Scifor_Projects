```python
#importing the libraries
import numpy as np
import gym
```

```python
#building the environment
env = gym.make('FrozenLake-v1', render_mode = 'human', new_step_api=True)
```

```python
#defining parameters
epsilon = 1.0
total_episodes = 5000
max_steps = 100
alpha = 0.1
gamma = 0.99
Q = np.zeros((env.observation_space.n, env.action_space.n))
```

```python
# Function to choose the next action using epsilon-greedy policy
def choose_action(state, Q, epsilon):
    if np.random.uniform(0, 1) < epsilon:
        return env.action_space.sample()
    else:
        return np.argmax(Q[state, :])

# Function to learn the Q value
def update(Q, state, state2, reward, action, action2, alpha, gamma):
    predict = Q[state, action]
    target = reward + gamma * Q[state2, action2]
    Q[state, action] = Q[state, action] + alpha * (target - predict)
```

```python
# SARSA learning
total_rewards = []
for episode in range(total_episodes):
    state1 = env.reset()
    action1 = choose_action(state1, Q, epsilon)
    t = 0
    episode_reward = 0

    while t < max_steps:
        env.render()
        result = env.step(action1)
        print(result)
        state2, reward, done, info = result[:4]
        #state2, reward, done, _ = env.step(action1)

        episode_reward += reward

        action2 = choose_action(state2, Q, epsilon)
        update(Q, state1, state2, reward, action1, action2, alpha, gamma)

        state1 = state2
        action1 = action2

        if done:
            break

        t += 1

    epsilon = max(0.1, epsilon * 0.995)  # Decay epsilon for exploration
    total_rewards.append(episode_reward)

    print(f"Episode {episode + 1} - Total Reward: {episode_reward}")
```

```python
# Evaluation
average_reward = sum(total_rewards) / total_episodes
print("Average performance:", average_reward)

# Visualizing the Q matrix
print(Q)
```
