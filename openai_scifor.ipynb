```python
def comp(PROMPT, MaxToken = 50, outputs = 3):
  #use openai's completion module
  response = openai.Completion.create(model = 'davinci-002', prompt = PROMPT, max_tokens = MaxToken, n = outputs) #model, user input, max num of tokens, num of outputs generated in one call
  output = list()
  for k in response['choices']:
    output.append(k['text'].strip())
  return output
PROMPT = """models that can be used for forecasting climate change"""
comp(PROMPT, MaxToken = 3000, outputs = 3)
```

```python
def get_embedding(text, model="text-embedding-ada-002"):
  return openai.Embedding.create(input=text,model=model)['data'][0]['embedding']
PROMPT = "What is Generative AI and give some Gen AI models"
get_embedding(PROMPT)
```

image generation

```python
import requests
from PIL import Image
from io import BytesIO
```

```python
def generate(text):
  res=openai.Image.create(prompt=text, n=1, size="256x256",)
  return res["data"][0]["url"]
text = "Image of an apartment"
url1 = generate(text)
response = requests.get(url1, stream=True)
Image.open(response.raw)
```
