# loading dataset, finding missing values

```python
df = pd.read_excel('P_Data_Extract_From_World_Development_Indicators.xlsx')
```

```python
df = df.dropna()
```

```python
df.replace('..', pd.NA, inplace=True)
```

```python
miss_values = df.isnull().sum()/len(df)*100
miss_values
```

```python
df.drop(columns = ['Country Code', 'Series Code'], inplace = True)
```

```python
cols_to_fill = miss_values[(miss_values >0)].index
df[cols_to_fill] = df[cols_to_fill].fillna(df[cols_to_fill].mean())
```

# filtering the values of population density and maternal mortality

```python
pd_mm_values = df[df['Series Name'].isin(['Population density (people per sq. km of land area)', 'Maternal mortality ratio (national estimate, per 100,000 live births)'])]
pd_mm_values
```

# filtering of population density and literacy rate in 2020, plotting the graph

```python
import matplotlib.pyplot as plt
pop_2020 = df[df['Series Name'] == 'Population density (people per sq. km of land area)']
literacy_2020 = df[df['Series Name'] == 'Literacy rate, adult total (% of people ages 15 and above)']
merged_data = pop_2020[['Country Name', '2020 [YR2020]']].merge(literacy_2020[['Country Name', '2020 [YR2020]']], on='Country Name')
merged_data.columns = ['Country Name', 'Population Density', 'Literacy Rate']

plt.figure(figsize=(10, 6))
plt.scatter(merged_data['Population Density'], merged_data['Literacy Rate'])
plt.xlabel('Population Density (2020)')
plt.ylabel('Literacy Rate, Adult Total (2020)')
plt.title('Population Density vs. Literacy Rate in 2020')
plt.grid(True)
plt.show()
```
